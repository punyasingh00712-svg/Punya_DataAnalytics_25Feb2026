# Student Dataset Cleaning using Pandas
This notebook identifies missing values, removes duplicates, replaces missing Attendance with mean, and replaces missing Marks with median.

```
import pandas as pd
import numpy as np
```

```
# Creating Student DataFrame
data = {
    'Student_ID': [1, 2, 3, 4, 5, 5],
    'Name': ['Aman', 'Bhavna', 'Chetan', 'Divya', 'Esha', 'Esha'],
    'Marks': [85, np.nan, 78, 90, np.nan, np.nan],
    'Attendance': [92, 88, np.nan, 95, 85, 85]
}

df = pd.DataFrame(data)
df
```

```
# Identify missing values
df.isnull().sum()
```

```
# Remove duplicate rows
df = df.drop_duplicates()
df
```

```
# Replace missing Attendance with mean
df['Attendance'].fillna(df['Attendance'].mean(), inplace=True)
df
```

```
# Replace missing Marks with median
df['Marks'].fillna(df['Marks'].median(), inplace=True)
df
```

```
# Final Cleaned Dataset
df
```
